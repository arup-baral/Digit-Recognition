```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from tensorflow import keras
```

```python
model = keras.models.load_model('digit_recognizer.h5')
model.summary()
```

```python
img = cv.imread('digit9.jpg')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
img_resized = cv.resize(img, (28,28))
img_resized = img_resized / 255
x = np.reshape(img_resized, [1,28,28,1])
x.shape
```

```python
plt.imshow(x[0])
```

```python
y = model.predict(x)
np.argmax(y)
```
